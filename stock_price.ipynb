```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options 
import os 
```

```python
# 크롤링할 웹 사이트 정보
base_url = 'https://finance.naver.com/sise/lastsearch2.nhn'


# get html
response = requests.get(base_url)


options = Options()
options.headless = True
browser = webdriver.Chrome(options=options)
browser.implicitly_wait(3)
browser.get(base_url)
time.sleep(5)
```

```python
# BeautifulSoup을 활용하여 데이터 파싱
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
lists = soup.findAll('div', {'class' : 'box_type_l'}) 
name = []
# print(lists)
# lists = soup.findAll('a')
for tmp in lists:
#     print(tmp.text)
    title_list = tmp.findAll('a', {'class': "tltle" })
    for tmp in title_list :
        #print(tmp.text)
        name.append(tmp.text)
print(name)
```
