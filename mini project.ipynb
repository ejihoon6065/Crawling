```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 크롤링할 웹 사이트 정보
base_url = 'https://www.tripadvisor.co.kr/'
seoul_url = '/Hotels-g294197-Seoul-Hotels.html'


# get html
response = requests.get(base_url+seoul_url)

# BeautifulSoup을 활용하여 데이터 파싱
soup = BeautifulSoup(response.text, 'html.parser')


#<div class="prw_rup prw_meta_hsx_listing_name listing-title" data-prwidget-name="meta_hsx_listing_name" data-prwidget-init=""><div class="listing_title"><a target="_blank" href="/Hotel_Review-g294197-d301817-Reviews-GRAND_WALKERHILL_Seoul-Seoul.html" id="property_301817" class="property_title prominent " data-clicksource="HotelName" onclick="return false;" dir="ltr">그랜드 워커힐 서울</a></div> </div>
# 맛집 목록 가져오기
lists = soup.findAll('div', {'class' : 'listing_title'}) 
hrefs = soup.find('a')
#<div class="ui_column is-narrow title_wrap"><a target="_blank" href="/Hotel_Review-g294197-d15068651-Reviews-L_Escape_Hotel-Seoul.html" id="property_15068651" class="property_title prominent " data-clicksource="HotelName" onclick="return false;" dir="ltr">레스케이프 호텔</a></div>
#<a target="_blank" href="/Hotel_Review-g294197-d15068651-Reviews-L_Escape_Hotel-Seoul.html" id="property_15068651" class="property_title prominent " data-clicksource="HotelName" onclick="return false;" dir="ltr">레스케이프 호텔</a>

# <div class="ui_column is-narrow title_wrap"><a target="_blank" href="/Hotel_Review-g294197-d15068651-Reviews-L_Escape_Hotel-Seoul.html" id="property_15068651" class="property_title prominent " data-clicksource="HotelName" onclick="return false;" dir="ltr">레스케이프 호텔</a></div>

print(lists)
```

```python
rest_id_list = []
rest_name_list = []
rest_url_list = []
names = []
full_urls = []
for item in lists:
   
    href = hrefs.get('href')
    full_url = base_url + href
    name = item.text

    names.append(name)
    full_urls.append(full_url)
print(len(names))
print(len(full_urls))    

        
result = pd.DataFrame(data={'[NAME]' : names, '[URL]' : full_urls})
result
    
 
```
